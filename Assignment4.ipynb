```python
#Assignment 4
#load the dictionary
def load_dic():
    dic_file = 'words.txt'
    dic = set()
    file = open(dic_file, 'r') if open(dic_file, 'r') else None
    for word in file:
        dic.add(word.strip().upper())
    file.close()
    return dic

#perform basic arithmetic operations
def perform_operation(o1, o2, operator):
    if operator == '+':
        return o1 + o2
    elif operator == '-':
        return o1 - o2
    elif operator == '*':
        return o1 * o2
    elif operator == '/':
        if o2 == 0:
            print("Division by zero error.")
            return None
        return o1 // o2
    else:
        print(f"Invalid operator: {operator}")
        return None
#evaluate expression without try
def evaluate_expression(expression, numbers):
    valid_operators = set("+-*/") 
    valid_numbers = set(map(str, numbers)) 
    tokens = expression.split()
    
    if len(tokens) % 2 == 0:
        print("Invalid expression format.")
        return None
    #initialize operand1 as the first number
    if tokens[0] not in valid_numbers:
        print(f"Invalid number: {tokens[0]}")
        return None
    
    result = int(tokens[0])  #start with the first number
    #the rest of the expression
    for i in range(1, len(tokens), 2):
        operator = tokens[i]
        if operator not in valid_operators:
            print(f"Invalid operator: {operator}")
            return None
        
        if tokens[i+1] not in valid_numbers:
            print(f"Invalid number: {tokens[i+1]}")
            return None
        
        o2 = int(tokens[i+1])
        result = perform_operation(result, o2, operator)
        
        if result is None: 
            return None
    
    return result

# Number Round
def numbers_round():
    print("=== Numbers Round ===")
    target = int(input("Please enter a three-digit target number: "))
    #input 6 numbers
    numbers = []
    print("Please enter 6 available numbers (between 1-10 or 25, 50, 75, 100):")
    for i in range(6):
        number = int(input(f"Enter number {i+1}: "))
        numbers.append(number)
    #input an arithmetic expression using those numbers to reach the target
    expression = input(f"Enter an arithmetic expression using these 6 numbers to get close to {target}: ")
    guess = evaluate_expression(expression, numbers)
    
    if guess is not None:
        print(f"Your result is: {guess}")
        #calculate the difference from the target
        difference = abs(target - guess)
        print(f"Difference from the target: {difference}")
        #calculate the score
        score = 100 - difference // 10
        print(f"Your score is: {score}")
    else:
        print("Invalid expression, cannot calculate result.")

# Letter Round
def letters_round(dic):
    print("=== Letters Round ===")
    vowels = "AEIOU"
    consonants = "BCDFGHJKLMNPQRSTVWXYZ"
    letters = [] 
    #choose vowels or consonants and input 10 letters
    for i in range(10):
        choice = input("Choose a vowel (V) or a consonant (C): ").upper()
        if choice == 'V':
            letter = input(f"Enter a vowel (A, E, I, O, U): ").upper()
            if letter in vowels:
                letters.append(letter)
            else:
                print("Invalid vowel input!")
                return
        elif choice == 'C':
            letter = input(f"Enter a consonant: ").upper()
            if letter in consonants:
                letters.append(letter)
            else:
                print("Invalid consonant input!")
                return
        else:
            print("Invalid choice, please enter V or C.")
            return
    print(f"Your chosen letters are: {letters}")
    #input the longest word they can form
    player_word = input("Enter the longest word you can form: ").upper()

    #check if the word is in the dictionary
    if player_word in dic:
        print(f"Your word '{player_word}' is valid!")
        longest_word_length = len(player_word)
    else:
        print(f"Your word '{player_word}' is not in the dictionary!")
        longest_word_length = 0
    #word length and score
    print(f"Your word is: {player_word}, and its length is: {longest_word_length}")
    score = 100 - (10 - longest_word_length) * 10
    print(f"Your score is: {score}")

#run the game
def play_game():
    print("Welcome to 'Des Chiffres et Des Lettres' game!")
    dic = load_dic()
    numbers_round()
    if dic:
        letters_round(dic)
    else:
        print("Cannot play the letters round without a valid dictionary.")

play_game()




```
